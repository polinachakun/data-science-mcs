# Task 2 - Pokemon

- a) Load the `pokemon.csv` dataset into a `pandas` `DataFrame`.
- b) Parse the `'abilities'` column into a list of strings.
- c) Use the [`explode`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.explode.html) method to create a new row for each ability of a Pokemon.
- d) Find the most common and rarest ability.
- e) Find the most common and rarest ability for each Generation.

Solution a)


```python
df_pokemon = pd.read_csv('pokemon.csv')
df_pokemon
```

Solution b)

```python
df_pokemon['abilities'] = df_pokemon['abilities'].apply(lambda x: x.strip('[]').replace("'", "").split(', '))
```

Solution c)

```python
df_exploded = df_pokemon.explode('abilities')
```

Solution d)

```python
most_common_ability = df_exploded['abilities'].value_counts().idxmax()
rarest_ability = df_exploded['abilities'].value_counts().idxmin()

print(f"Most common ability: {most_common_ability}")
print(f"Rarest ability: {rarest_ability}")
```

Solution e)

```python
# Assuming 'generation' is a column in the dataset that indicates the Pokemon's generation.
generation_ability_counts = df_exploded.groupby('generation')['abilities'].value_counts()

# Finding the most common and rarest ability per generation
most_common_per_gen = generation_ability_counts.groupby('generation').idxmax()
rarest_per_gen = generation_ability_counts.groupby('generation').idxmin()

print("\nMost common ability for each Generation:")
print(most_common_per_gen)

print("\nRarest ability for each Generation:")
print(rarest_per_gen)
```
